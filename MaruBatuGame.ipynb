```python
import re

class MaruBatuGame:
    def __init__(self):
        self.table = [["-","-","-"],["-","-","-"],["-","-","-"]]
        self.turn = 0

    def show(self):
        print("  A B C")
        print("1 " + self.table[0][0] + " " + self.table[0][1] + " " + self.table[0][2])
        print("2 " + self.table[1][0] + " " + self.table[1][1] + " " + self.table[1][2])
        print("3 " + self.table[2][0] + " " + self.table[2][1] + " " + self.table[2][2])

    def inputRow(self):
        print("Please choose row 1 or 2 or 3")
        self.row = input()
        if re.match("^1$|^2$|^3$",self.row) == None:
            return False
        else:
            return True

    def inputCol(self):
        print("Please choose col A or B or C")
        self.col = input()
        if re.match("^a$|^b$|^c$|^A$|^B$|^C$",self.col) == None:
            return False
        else:
            return True

    def put(self):
        if(self.col == "A" or self.col == "a"):
            if(self.row == "1" and self.table[0][0] == "-"):
                self.table[0][0] = "o" if self.turn == 0 else "x"
                return True
            if(self.row == "2" and self.table[1][0] == "-"):
                self.table[1][0] = "o" if self.turn == 0 else "x"
                return True
            if(self.row == "3" and self.table[2][0] == "-"):
                self.table[2][0] = "o" if self.turn == 0 else "x"
                return True
        if(self.col == "B" or self.col == "b"):
            if(self.row == "1" and self.table[0][1] == "-"):
                self.table[0][1] = "o" if self.turn == 0 else "x"
                return True
            if(self.row == "2" and self.table[1][1] == "-"):
                self.table[1][1] = "o" if self.turn == 0 else "x"
                return True
            if(self.row == "3" and self.table[2][1] == "-"):
                self.table[2][1] = "o" if self.turn == 0 else "x"
                return True
        if(self.col == "C" or self.col == "c"):
            if(self.row == "1" and self.table[0][2] == "-"):
                self.table[0][2] = "o" if self.turn == 0 else "x"
                return True
            if(self.row == "2" and self.table[1][2] == "-"):
                self.table[1][2] = "o" if self.turn == 0 else "x"
                return True
            if(self.row == "3" and self.table[2][2] == "-"):
                self.table[2][2] = "o" if self.turn == 0 else "x"
                return True
        return False

    def isFinished(self):
        if(self.table[0][0] == self.table[0][1] and self.table[0][0] == self.table[0][2] and self.table[0][0] != "-"):
            return True
        if(self.table[1][0] == self.table[1][1] and self.table[1][0] == self.table[1][2] and self.table[1][0] != "-"):
            return True
        if(self.table[2][0] == self.table[2][1] and self.table[2][0] == self.table[2][2] and self.table[2][0] != "-"):
            return True
        if(self.table[0][0] == self.table[1][0] and self.table[0][0] == self.table[2][0] and self.table[0][0] != "-"):
            return True
        if(self.table[0][1] == self.table[1][1] and self.table[0][1] == self.table[2][1] and self.table[0][1] != "-"):
            return True
        if(self.table[0][2] == self.table[1][2] and self.table[0][2] == self.table[2][2] and self.table[0][2] != "-"):
            return True
        if(self.table[0][0] == self.table[1][1] and self.table[0][0] == self.table[2][2] and self.table[0][0] != "-"):
            return True
        if(self.table[0][2] == self.table[1][1] and self.table[1][1] == self.table[2][0] and self.table[0][2] != "-"):
            return True
        return False

    def play(self):
        while self.isFinished() == False:
            self.show()
            print("Player","o" if self.turn == 0 else "x",sep=":")
            while self.inputRow() == False :
                print("You imputed incorrect character! Please imput '1' or '2' or '3'.")
            while self.inputCol() == False:
                print("You imputed incorrect character! Please imput 'A' or 'B' or 'C'.")
            if(self.put() == False):
                print("This position already has been put.")
                continue
            self.turn = 1 if self.turn == 0 else 0
        self.show()
        if(self.turn != 1):
            print("x has won!!")
        else:
            print("o has won!!")

maruBatuGame = MaruBatuGame()
maruBatuGame.play()
```
